```python
status_home  = []

hg = home['home_team_goal_count'].tolist()
ag = home['away_team_goal_count'].tolist()


for i, hgx in enumerate(hg):
    agx = ag[i]
    if hgx > agx:
        status = 'Win_home'
        status_home.append(status)
    elif hgx < agx:
        status_l = 'Loss_home'
        status_home.append(status_l)
    else :
        status_d = 'Draw_home'
        status_home.append(status_d)
        
```

```python
home.insert(11, "Status_home", status_home, True)
y = home['Status_home'].tolist()
print(y)
```

```python
status_away  = []

hg = away['home_team_goal_count'].tolist()
ag = away['away_team_goal_count'].tolist()


for i, hgx in enumerate(hg):
    agx = ag[i]
    if hgx > agx:
        status = 'Win_away'
        status_away.append(status)
    elif hgx < agx:
        status_l = 'Loss_away'
        status_away.append(status_l)
    else :
        status_d = 'Draw_away'
        status_away.append(status_d)
```

```python
away.insert(11, "Status_away", status_away, True)
```

```python
x = away['Status_away'].tolist()
print(x)
```

```python
import matplotlib.pyplot as plt
   
year = away['Game Week'].tolist()
unemployment_rate = away['Status_away'].tolist()
  
plt.plot(year, unemployment_rate)
plt.title('Everton Home Team Goals')
plt.xlabel('Game week')
plt.ylabel('status_away')
plt.show()

```
