```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
df = pd.read_csv('Teams1.csv')
teams_a = df['name'].tolist()
goals_a = df['league_position_away'].tolist()
```

```python
teams_home = df['name'].tolist()
goals_home = df['wins_away'].tolist()

fig = plt.figure(figsize = (10, 5))


# creating the bar plot
plt.bar(teams_home, goals_home, color ='maroon',
        width = 0.4)
 
plt.xlabel("EPL Teams 2223")
plt.ylabel("No. of goals")
plt.title("goals_conceded_away")
plt.show()
```
